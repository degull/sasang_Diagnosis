이미지를 224x224 크기로 crop & 얼굴만 출력되도록


```python
import os
import cv2

def crop_face(image_path, output_path):
    image = cv2.imread(image_path)


    if image is None:
        print(f"Error loading image: {image_path}")
        return

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    if len(faces) == 0:
        return

    face = faces[0]
    x, y, w, h = face

    # 이미지 crop
    face_image = image[y:y + h, x:x + w]

    # 224x224
    face_image = cv2.resize(face_image, (224, 224))

    # 이미지 저장
    cv2.imwrite(output_path, face_image)

def crop_body(image_path, output_path):
    image = cv2.imread(image_path)

    if image is None:
        print(f"Error loading image: {image_path}")
        return

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    body_cascade = cv2.CascadeClassifier('haarcascade_fullbody.xml')
    bodies = body_cascade.detectMultiScale(gray, 1.1, 4)

    if len(bodies) == 0:
        return

    body = bodies[0]
    x, y, w, h = body

    body_image = image[y:y + h, x:x + w]

    body_image = cv2.resize(body_image, (224, 224))

    cv2.imwrite(output_path, body_image)

def main():
    input_dir = 'image'
    output_dir = 'output_images'

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for image_name in os.listdir(input_dir):
        image_path = os.path.join(input_dir, image_name)

        crop_face(image_path, os.path.join(output_dir, 'face_' + image_name))
        crop_body(image_path, os.path.join(output_dir, 'body_' + image_name))

if __name__ == '__main__':
    main()
```

```python
import os
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt

def crop_face(image_path, output_path):
    image = cv2.imread(image_path)


    if image is None:
        print(f"Error loading image: {image_path}")
        return

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    if len(faces) == 0:
        return

    face = faces[0]
    x, y, w, h = face

    # 이미지 crop
    face_image = image[y:y + h, x:x + w]

    # 224x224
    face_image = cv2.resize(face_image, (224, 224))

    # 이미지 저장
    cv2.imwrite(output_path, face_image)

def main():
    input_dir = 'input_image'   # 파일명
    output_dir = 'output_images'

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    image_names = os.listdir(input_dir)
    total_images = len(image_names)

    # Progress bar setup
    progress_bar = tqdm(total=total_images, desc='Processing images', unit='image')

    for image_name in image_names:
        image_path = os.path.join(input_dir, image_name)

        crop_face(image_path, os.path.join(output_dir, 'face_' + image_name))
        #crop_body(image_path, os.path.join(output_dir, 'body_' + image_name))

        # Update progress bar
        progress_bar.update(1)

    # Close progress bar
    progress_bar.close()

    # Plot the progress
    plt.figure(figsize=(8, 6))
    plt.bar(['Processed', 'Remaining'], [total_images, 0], color=['green', 'gray'])
    plt.xlabel('Status')
    plt.ylabel('Number of Images')
    plt.title('Image Processing Progress')
    plt.show()

if __name__ == '__main__':
    main()
```

# 이미지 이름명 변경

```python
import os

# 이미지 파일이 들어있는 디렉토리 경로 지정
directory_path = '/content/name'

# 디렉토리 내의 모든 파일 목록 가져오기
file_list = os.listdir(directory_path)

# 이미지 파일만 필터링하여 정렬
image_files = [file for file in file_list if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
image_files.sort()

# 이미지 파일 이름 변경
for idx, old_name in enumerate(image_files, start=212):
    # 새로운 파일 이름 생성 (예: 1.jpg, 2.png, ...)
    new_name = f"{idx}.{old_name.split('.')[-1]}"

    # 파일 이름 변경
    old_path = os.path.join(directory_path, old_name)
    new_path = os.path.join(directory_path, new_name)
    os.rename(old_path, new_path)

print("이미지 파일 이름이 성공적으로 변경되었습니다.")
```
